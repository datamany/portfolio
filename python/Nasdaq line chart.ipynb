Create function that is going to return a comparison plot of daily closing price on Nasdaq by Google, Apple, Amazon and Microsoft stocks.
Note: names of arguments must be automatically defined as legend names.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
goog = pd.read_csv('HistoricalQuotesGoog5years.csv', index_col = "date", skiprows=[1])
aapl = pd.read_csv('HistoricalQuotesAapl5years.csv', index_col = "date", skiprows=[1])
amzf = pd.read_csv('HistoricalQuotesAMZF5years.csv', index_col = "date", skiprows = [1])
msft = pd.read_csv('HistoricalQuotesMSFT5years.csv', index_col = "date", skiprows = [1])
```

```python
def sorting(*args):
    for i in args:
        i.sort_values(by=['date'], ascending = True, inplace = True)
sorting(goog,aapl,amzf,msft)
```

```python
def plot_with_legend1 (*args,comparison_column= "close", w=15,h=5):
    def get_df_name(df):
        name =[x for x in globals() if globals()[x] is df][0]
        return name
    legendlist =[]
    for i in args:
        a = get_df_name(i)
        legendlist.append(a)
        i[comparison_column].plot(figsize= (w,h))
    plt.legend(legendlist,bbox_to_anchor=(1, 1), loc='upper left', ncol=1)
    plt.xticks(rotation =45)
```

```python
plot_with_legend1(aapl,goog,msft,amzf)
```
