```python
import pandas as pd; import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_extracted_df_filtered.csv')
```

```python
df['state'] = df['Location'].apply(lambda x: x.split(',')[1].strip() if len(x.split(','))>1 else  None)
df['Location'] = df['Location'].apply(lambda x: x.split(',')[0].strip())
```

```python
df[['Location', 'state']]
```

```python
df.GDJobTitle.value_counts().head(20).sort_values(ascending = True).plot(kind = 'barh', figsize = (8,10))
plt.grid(axis = 'x')
plt.xticks( fontsize = '14')
plt.yticks(fontsize = '14')
plt.show()
```

```python
## Location
df.Location.value_counts().head(20).sort_values(ascending = True).plot(kind = 'barh', figsize = (8,10))
plt.grid(axis = 'x')
plt.xticks( fontsize = '14')
plt.yticks(fontsize = '14')
plt.show()
```

```python
df.state.value_counts().head(20).sort_values(ascending = False).plot(kind = 'bar', figsize = (20,4))
plt.grid(axis = 'y')
plt.xticks(rotation = '45', fontsize = '14')
plt.yticks(fontsize = '14')
plt.show()
```
